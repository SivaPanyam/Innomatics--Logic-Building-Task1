
### Problem 1: Employee Performance Bonus Eligibility

```python


#%%
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Get the highest score
highest_score = max(employees.values())

# Collect all employees with the highest score (handles ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

## Problem 2: Search Query Keyword Analysis

```python
import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans("", "", string.punctuation))

# Count keyword frequency
freq = {}
for word in query.split():
    freq[word] = freq.get(word, 0) + 1

# Keep only keywords with frequency > 1
result = {word: count for word, count in freq.items() if count > 1}

print(result)
```

## Problem 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Keep only even readings and store as (hour_index, reading_value)
valid_readings = [
    (hour, value)
    for hour, value in enumerate(sensor_readings)
    if value % 2 == 0
]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Count users per domain
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate and display percentage usage
total_users = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

## Problem 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
average_sales = sum(sales) / len(sales)

# 30% above average threshold (rounded for reporting)
threshold = round(average_sales * 1.30, -2)

# Display days where sales cross the spike threshold
for day, value in enumerate(sales, start=1):
    if value >= threshold:
        print(f"Day {day}: {value}")
```

## Problem 6: Duplicate User ID Detection

```python
from collections import Counter

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences of each user ID
counts = Counter(user_ids)

# Print only duplicates (count > 1)
for user_id, count in counts.items():
    if count > 1:
        print(f"{user_id} → {count} times")
```
